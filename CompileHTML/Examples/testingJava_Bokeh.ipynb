```python
import numpy as np
from bokeh.models import LinearAxis, Range1d, DataRange1d
from bokeh.layouts import row, widgetbox
from bokeh.models import CustomJS, Slider, CheckboxGroup
from bokeh.plotting import figure, output_file, show, ColumnDataSource
from IPython.display import IFrame
import matplotlib.pyplot as plt
from bokeh.io import output_notebook
output_notebook()
```

```python

x = np.linspace(0, 10, 500)
y = np.sin(x)

source = ColumnDataSource(data=dict(x=x, y=y, opt1 = x, opt2=y))

plot = figure(y_range=(-10, 10), plot_width=400, plot_height=400)

line1 = plot.line('x', 'y', source=source, line_width=3, line_alpha=0.6, line_color = 'green')

callback = CustomJS(args={}, code="""
    //xrange=plot.x_range
    data = source.data
    console.log(plot1.x_range.end)
    if (0 in checkbox.active) {
        //line1.visible = true
        line1.glyph.line_color = 'blue'
        data['y'] = data['opt1']
    } else {
        //line1.visible = false
        line1.glyph.line_color = 'red'
        data['y'] = data['opt2']
    }
    
""")

checkbox = CheckboxGroup(labels=["Line 0"], active=[0, 1, 2], callback=callback, width=100)
callback.args = dict(line1=line1, checkbox=checkbox, source=source, plot1 = plot)
layout = row(
    plot,
    widgetbox(checkbox),
)

##output_file("slider3.html", title="slider.py example")

show(layout)
```
