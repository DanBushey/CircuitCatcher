```python
from bokeh.io import show
from bokeh.layouts import column
from bokeh.models import Button, CustomJS
from bokeh.plotting import figure
from bokeh.io import output_notebook, show
from bokeh.models.renderers import GlyphRenderer
import numpy as np
output_notebook()

p = figure(tools="reset,pan,wheel_zoom,lasso_select")
t= p.circle(list(range(10)), list(range(10)), name='c')

b = Button()
b.js_on_click(CustomJS(args=dict(p=p), code="""
    p.reset.emit()
"""))
glyph_name_list = ['c']
renderers = p.select(dict(type=GlyphRenderer))
for r in renderers:
    if r.name in glyph_name_list:
        col = r.glyph.y
        r.data_source.data[col] = [np.nan] * len(r.data_source.data[col])

show(column(p, b))
```
