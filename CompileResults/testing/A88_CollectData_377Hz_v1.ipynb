```python
import pandas as pd
from pathlib import Path
import numpy as np


```

```python
#load the summary xlsx sheet 
output = str(Path.cwd().parents[0] / 'A88_Data4_377Hz_511.pckl')
path = str(Path.cwd().parents[0] / 'A88_Summary_data4_377Hz_511.xlsx')
xls = pd.read_excel(path)
keep_columns = ['No.', 'Sample Name', 'Genotype', 'Date', 'Path', 'L2_Columns', 'Wavelength', 'Magnification', 'size', 'Pockels', 'PMT2', 'Filter']
xls = xls[keep_columns]
```

```python
#adding signal data to each row in xls file
#file_name = 'SummaryData.hd5'
data = []
for row, dseries in xls.iterrows():
    #row = 0
    #dseries = xls.iloc[row]
    targetfolder = Path(dseries['Path'])
    targetfile = targetfolder / str('CROI_' + targetfolder.parts[-1] + '.pckl')
    data.append( pd.read_pickle(targetfile))

```

```python
#organize Time, Signal and Voltage so they sync with first light ons and include 10 stimulation periods
time_sync = []
l2_sync = []
back_sync = []
voltage_sync = []

for cdata in data:
    l2_sync.append(cdata.loc['L2'].values)
    back_sync.append(cdata.loc['Background'].values)
    voltage_sync.append(cdata.loc['voltage'].values)
    time_sync.append(cdata.columns)
    
        
xls['Time_sync'] = time_sync
xls['L2_sync'] = l2_sync
xls['Back_sync'] = back_sync
xls['Voltage_sync'] = voltage_sync
```

```python
#save to pickel file

xls.to_pickle(output)
```
